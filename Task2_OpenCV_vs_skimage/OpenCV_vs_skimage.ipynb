# üì∑ OpenCV vs scikit-image: Two Lenses on Images  

As part of my Huawei AI Scholarship journey, I wanted to compare two popular Python libraries for image processing:  
- **OpenCV** ‚Üí fast, powerful, widely used in industry.  
- **scikit-image (skimage)** ‚Üí clean, NumPy-based, great for research.  


1Ô∏è‚É£ Theory: Comparing scikit-image (skimage) vs OpenCV


| Aspect                | **OpenCV**                                                                     | **scikit-image (skimage)**                                             |
| --------------------- | ------------------------------------------------------------------------------ | ---------------------------------------------------------------------- |
| **Purpose**           | General-purpose **computer vision library** (C++ backend, Python bindings).    | Pure Python library for **image processing & analysis**.               |
| **Performance**       | Optimized in C/C++ ‚Üí very fast, real-time capable.                             | Python + NumPy ‚Üí slower, not ideal for real-time.                      |
| **Functions**         | Covers image processing, feature detection, object recognition, video I/O, ML. | Focuses on image processing: filtering, segmentation, transformations. |
| **Ease of Use**       | API sometimes low-level, many functions require manual tuning.                 | Higher-level, user-friendly, works naturally with NumPy arrays.        |
| **Installation Size** | Bigger dependency, includes video, deep learning support.                      | Lighter, focused only on images.                                       |
| **Best For**          | Industry, production, robotics, real-time CV.                                  | Research, prototyping, teaching, scientific computing.                 |


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, filters

# Load image
img = io.imread("/content/Dog_face.png")  # free image
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

## 1Ô∏è‚É£ Grayscale Conversion
Both libraries can convert to grayscale. Let's compare.


```python
# OpenCV grayscale
gray_cv = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# skimage grayscale
gray_sk = color.rgb2gray(img)  # result is float [0,1]

# Show side by side
plt.figure(figsize=(10,5))
plt.subplot(1,2,1), plt.imshow(gray_cv, cmap='gray'), plt.title("OpenCV Grayscale"), plt.axis("off")
plt.subplot(1,2,2), plt.imshow(gray_sk, cmap='gray'), plt.title("skimage Grayscale"), plt.axis("off")
plt.show()
```

## 2Ô∏è‚É£ Edge Detection
- **OpenCV** ‚Üí Canny detector.  
- **skimage** ‚Üí Sobel filter.  


```python
# OpenCV Canny
edges_cv = cv2.Canny(gray_cv, 100, 200)

# skimage Sobel
edges_sk = filters.sobel(gray_sk)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1), plt.imshow(edges_cv, cmap='gray'), plt.title("OpenCV Canny"), plt.axis("off")
plt.subplot(1,2,2), plt.imshow(edges_sk, cmap='gray'), plt.title("skimage Sobel"), plt.axis("off")
plt.show()
```

## 3Ô∏è‚É£ Thresholding
- **OpenCV** ‚Üí `cv2.threshold`  
- **skimage** ‚Üí Otsu's method  


```python
# OpenCV threshold
_, th_cv = cv2.threshold(gray_cv, 128, 255, cv2.THRESH_BINARY)

# skimage Otsu threshold
from skimage.filters import threshold_otsu
thresh_val = threshold_otsu(gray_sk)
th_sk = gray_sk > thresh_val

plt.figure(figsize=(12,5))
plt.subplot(1,2,1), plt.imshow(th_cv, cmap='gray'), plt.title("OpenCV Threshold"), plt.axis("off")
plt.subplot(1,2,2), plt.imshow(th_sk, cmap='gray'), plt.title("skimage Otsu"), plt.axis("off")
plt.show()
```

## 4Ô∏è‚É£ Reflection: OpenCV vs skimage  

| Feature | OpenCV | skimage |
|---------|--------|----------|
| Speed | ‚ö° Fast, C++ optimized | üê¢ Slower, Python/NumPy |
| Scope | Wide: video, ML, vision pipelines | Narrow: scientific image analysis |
| Ease | More boilerplate | Very Pythonic |
| Use Case | Robotics, real-time | Research, prototyping |

**Conclusion:**  
- If I need **speed & production**, I‚Äôll use OpenCV.  
- If I need **clarity & experimentation**, I‚Äôll use skimage.  

This comparison gave me a better sense of how to pick the right tool depending on my project.
